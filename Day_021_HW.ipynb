```python
# 作業：取得另一個 dataset：titanic
# 導入必要的程式庫
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 

# 取得 Titanic 資料集
df = sns.load_dataset('titanic')
df.info()

# 題目1.做箱形圖
sns.boxplot(data = df, orient = "w")
plt.show()
```

```python
# 題目2.利用 facet grid 繪圖並分析
sns.set(style='whitegrid') 
g = sns.FacetGrid(df, col='survived') 
g.map(plt.hist, 'sex')
# 從圖片性別跟生存率的關係，可以發現女性的存活比例高出男性許多
h = sns.FacetGrid(df, col='survived') 
h.map(plt.hist, 'age')
# 從圖片年齡跟生存率的關係，可以發現20~40歲的存活比例最高
```

```python
# 題目3.繪製小提琴圖
sns.set(style='whitegrid') 
sns.violinplot(x='sex',y='survived',hue='who',data=df)
# 從圖片性別跟生存率的關係，可以發現女性與小孩的存活比例高出男性許多
```
