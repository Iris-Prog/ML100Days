```python
# 題目1. 繪製模型殘差圖型
import numpy as np 
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 

#設定圖形樣式 - whitegrid
sns.set(style="whitegrid")

# 利用 NUMPY 去建立資料集
# np.random.RandomState 設定數學式
rs = np.random.RandomState(0)
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)

# 繪製模型殘差圖型
sns.residplot(x=x, y=y, lowess=True)
```

```python
# 題目2.(1)使用 distplot()使用簡單的規則來正確猜測預設情況下正確的數位, 
#          但嘗試更多或更少的 bin 可能會顯示資料中的其他特徵 
# ※ 練習時已有 distplot 即將被棄用的警示, 故以下改用 displot
sns.set(color_codes=True)
x = np.random.normal(size=100) 
#sns.distplot(x);
#sns.distplot(x, bins=20, kde=False, rug=True);
sns.displot(x)
sns.displot(x, bins=10, kde=False, rug=True)
sns.displot(x, bins=20, kde=False, rug=True)
sns.displot(x, bins=50, kde=False, rug=True)
```

```python
# 題目2.(2)有無kde對圖形分布的影響
sns.displot(x, bins=20, kde=False, rug=True)
sns.displot(x, bins=20, kde=True, rug=True)
# 答案:2.(2)有無 kde 對圖形分布無影響，但有 kde 圖形會顯示密度曲線，助於資料判讀
```
