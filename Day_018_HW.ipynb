```python
# 作業1. 畫出 cos 圖檔，並儲存
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0,180)
y_cos = np.cos(x * np.pi / 180.0)
plt.plot(x, y_cos)
plt.savefig("y_cps.png",dpi=300,format="png")
```

```python
#作業 2. 給定散點圖顏色
n = 1024
X = np.random.normal(0,1,n)
Y = np.random.normal(0,1,n)
colors = np.random.rand(n)
area = np.random.rand(n)*120
 
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.scatter(X, Y, s=area, c=colors, alpha=0.5)
plt.title("Scatter plot")
plt.show()
```
